## 1. Bruit Gaussien (Distribution Normale)
ModÃĐlise le bruit thermique ou les erreurs de mesure (AWGN).

### ð `np.random.normal(loc, scale, size)`
* **UtilitÃĐ :** GÃĐnÃĻre des ÃĐchantillons suivant une loi normale.

* **Ce qu'elle retourne :** Un array de flottants. Exemple : `np.random.normal(0.0, 2.0, 1000)`

```python
# GÃĐnÃĻre un vecteur de 1000 ÃĐchantillons de bruit blanc (moyenne 0, ÃĐcart-type 2)
bruit = np.random.normal(loc=0.0, scale=2.0, size=1000)
```

* **UtilitÃĐ :** Nombres flottants avec probabilitÃĐ ÃĐgale sur tout l'intervalle.
* **ParamÃĻtres clÃĐs :**
    * `low` (float) : Borne infÃĐrieure (incluse).
